```python
import math
import itertools
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

```

```python
#Parameter

M = 10 # number of simulations
t = 100 # time


```

```python
random_walk = [-1, 1]
steps = np.random.choice(random_walk, size=(M,t)).T
origine = np.zeros((1,M))
rw_paths = np.concatenate([origine, steps]).cumsum(axis=0)
```

```python
plt.plot(rw_paths)
plt.xlabel("Years (t)")
plt.ylabel("Move")
plt.show()
```
